# Task 2 : Data Visualization Monthly Hotel Booking Analysis Based on Hotel Type

```python
df.head()
```

```python
# jumlah pemesanan nya berdasarkan jenis hotel dan bulannya, count jumlah booking perbulannya, nunique -> untuk hitung nilai unik yearnya.
# Ternyata nilai unik tahunnya tiap bulan berbeda, untuk September dan Oktober datanya dari 3 tahun.
df2=df.groupby (['hotel','arrival_date_month'])['arrival_date_year'].agg({'nunique','count'}).reset_index()
df2
```

```python
#kita hitung rata-rata bookiingnya per bulan 
df2['rata_rata_booking']= round(df2['count']/df2['nunique'])
df2
```

```python
mon = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
df2['month'] = df2['arrival_date_month'].str[:3]
df2['month'] = pd.Categorical(df2['month'],categories=mon ,ordered = True)
df2
#x = month, hue = hotel(tipe),  y = rata_rata_booking, 
#seaborn untuk plot
```

```python
plt.figure(figsize=(10, 8))
plt.title('Grafik Rata-rata Booking Hotel perbulan Berdasarkan Jenis Hotel', fontsize = 18);
sns.lineplot(x="month", y ='rata_rata_booking', hue = 'hotel', data=df2)
plt.text(1,4000, "Rata-rata booking tertinggi :",fontsize=13,weight= 'normal',color='#7393B3');
plt.text(1,3890, "4116 pesanan",fontsize=13,weight= 'normal',color='#7393B3');
plt.legend(loc = 2, bbox_to_anchor=(1,1));
plt.axvline('Jul', color='#89CFF0', linestyle='dotted',ymin=0, ymax=0.95)
plt.axvline('Dec', color='#6495ED', linestyle='dotted',ymin=0, ymax=0.86)
plt.axvline('Jun', color='#FFA500', linestyle='dotted',ymin=0, ymax=0.26)
plt.axvline('Dec', color='#FFA500', linestyle='dotted',ymin=0, ymax=0.25)
plt.tight_layout()
```

#Task 4 : Impact Analysis of Lead Time on Hotel Bookings Cancellation Rate

```python
df.info()
```

```python
df.lead_time.value_counts()
```

```python
df['Cat_Leadtime']=np.where(df['lead_time']<=7,'1 Minggu',
                            np.where(df['lead_time']<=30,'1 Bulan',
                                     np.where(df['lead_time']<=60,'2 Bulan',
                                              np.where(df['lead_time']<=90,'3 Bulan',
                                                        np.where(df['lead_time']<=180,'6 Bulan',
                                                                np.where(df['lead_time']<=365,'1 Tahun',
                                                                                    '> 1 tahun')
                                                                )
                                                        )
                                              )
                                      )
                            )
df.Cat_Leadtime.value_counts()
```

```python
df4a=df.groupby (['hotel','Cat_Leadtime'])['Cat_Leadtime'].agg({'count'}).reset_index()
df4a = df4a.rename(columns ={'count':'jumlah_reservasi'})
df4a
```

```python
df4 = df[df['reservation_status']=='Canceled'] # difilter yang Cancelled saja
df4
```

```python
df4b=df4.groupby (['hotel','Cat_Leadtime'])['Cat_Leadtime'].agg({'count'}).reset_index()
df4b = df4b.rename(columns ={'count':'jumlah_reservasi_cancelled'})
df4b
```

```python
df4_total= df4a.merge(df4b,left_on=['hotel','Cat_Leadtime'],right_on=['hotel','Cat_Leadtime'], how ='left')
df4_total
df4_total['ratio_cancel_lead']=df4_total['jumlah_reservasi_cancelled']/df4_total['jumlah_reservasi']*100
df4_total
```

```python
interval_urut = ['1 Minggu','1 Bulan','2 Bulan','3 Bulan','6 Bulan','1 Tahun','> 1 tahun']
df4_total['month'] = pd.Categorical(df4_total['Cat_Leadtime'],categories=interval_urut ,ordered = True)
df4_total
```

```python
plt.figure(figsize=(10, 6))
plt.title('Grafik Canceled rate Booking Hotel terhadap Lead Time Berdasarkan Jenis Hotel', fontsize = 18);
sns.lineplot(x="month", y ='ratio_cancel_lead', hue = 'hotel', data=df4_total)
plt.legend(loc = 2, bbox_to_anchor=(1,1));
plt.axvline('> 1 tahun', color='#89CFF0', linestyle='dotted',ymin=0, ymax=0.95)
plt.tight_layout()
```

## Recap Task 4
Hal yang sudah dilakukan:
- membuat kategori baru untuk lead time dengan menggunakan np.where
- menghitung jumlah reservasi cancel dan seluruhnya per kategori, dan dari masing-masing jenis hotel
- membuat visualisasi

Interpretasi:
- Dari masing-masing kategori lead time, semakin lama lead time nya semakin tinggi pula ratio cancelnya. Hal ini berlaku untuk kedua jenis hotel. 
- Cancel ratio paling tinggi terjadi pada pelanggan dengan lead time > 1 tahun, yakni > dari 70% pada city hotel, dan 45% pada resort hotel.
